### Lambda Functions in python

lambda function are small anonymous functions defined using the lambda keyword.They can have any number of arguments but only one expression.They are commonly used for short operations or as arguments to higher-order fucntions.

```python
#Syntax
lambda arguments: expression
```

```python
def addition(a,b):
  return a+b
```

```python
addition(2,3)
```

```python
add = lambda a,b:a+b
type(add)
print(add(5,6))
```

```python
def even(num):
  if num%2 == 0:
    return True
even(16)
```

```python
even = lambda num:num%2==0
even(9)
```

```python
def addition(a,b,c):
  return a+b+c
addition(2,3,4)
```

```python
addition = lambda a,b,c:a+b+c
addition(5,5,5)
```

```python
## map()- applies a function to all items in a list
numbers=[1,2,3,4,5,6]
def sqare(number):
  return number**2

sqare(2)
```

```python
list(map(lambda x:x**2,numbers)) ## map object --> map(func,*iterables)
```
